### Keyboard shortcuts we will use
To use **cell management shortcuts**, first click a cell and then press **Esc** to exit editing mode.

- **Shift + Enter** → run the current cell and move to the next cell
- **Ctrl + Enter** → run the current cell and stay on the same cell
- **A** → add a new cell above
- **B** → add a new cell below
- **M** → convert the selected cell to Markdown
- **Y** → convert the selected cell back to Code

## Task 9 – Restart the runtime
Sometimes in Colab, variables stay in memory after you run cells.  
To see that clearly, try this:

1. Run the code cell below
2. Then go to the menu and choose:

**Runtime → Restart runtime**

3. After the restart, try running only the **second** code cell in this section
4. See what happens

````python
message = "This variable exists only after this cell is run."
print(message)
````

````python
print(message)
````

### What is the point of this exercise?
This shows that Colab keeps information in memory **only while the runtime is active**.

After a restart:
- variables are cleared
- previous results may still be visible on screen
- but the notebook memory is reset

That is why restarting and rerunning cells is an important part of working correctly.
