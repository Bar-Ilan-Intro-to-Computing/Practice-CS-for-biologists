## Before you start

Keep these distinctions in mind:

- A **list** keeps order and allows duplicates.
- A **set** removes duplicates and is useful for membership and overlap.
- A **tuple** is ordered but immutable.
- A **dictionary** maps **keys** to **values**.

## 2) Sets: unique elements and overlap between experiments

Sets are especially useful when:
- order is not important,
- duplicates should be ignored,
- we want overlap, difference, or union.


```python
exp_a = {"TP53", "EGFR", "MYC", "KRAS", "BRAF"}
exp_b = {"EGFR", "BRAF", "PIK3CA", "PTEN", "MYC"}

print("shared genes      ->", exp_a & exp_b)
print("all detected      ->", exp_a | exp_b)
print("only in exp_a     ->", exp_a - exp_b)
print("only in exp_b     ->", exp_b - exp_a)
```

### Predict before running

If we convert the following list to a set:
- how many unique values will remain?
- will the original order be preserved?


```python
mutations = ["A>T", "A>T", "G>C", "del", "ins", "G>C", "A>T"]
unique_mutations = set(mutations)

print("original list:", mutations)
print("as a set     :", unique_mutations)
print("number of unique mutation types:", len(unique_mutations))
```

## 3) Dictionaries: mapping one thing to another

A dictionary is useful when we want to answer questions such as:
- What is the condition of sample `S4`?
- What is the length of gene `TP53`?
- What is the count of mutation type `del`?


```python
gene_length = {
    "TP53": 1179,
    "EGFR": 3636,
    "BRCA1": 5589,
    "MYC": 1323
}

print("Length of EGFR:", gene_length["EGFR"])
print("All genes     :", list(gene_length.keys()))
print("All lengths   :", list(gene_length.values()))
```

### Safe dictionary access

Direct access like `d[key]` fails if the key does not exist.  
Sometimes `get(...)` is safer.


```python
print(gene_length.get("TP53"))
print(gene_length.get("ALK"))
print(gene_length.get("ALK", "not found"))
```
